<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h1 style="font-size: 40px; margin-bottom: 14px; color: #2c3e50;">SECTION 1: Introduction to PDL</h1><h2 style="font-size: 31px; color: #34495e; margin-bottom: 20px;">1.1: Why Regular String Prompting Breaks</h2><p style="font-size: 17px; color: #7f8c8d; line-height: 1.75; margin-bottom: 14px;">String prompting is fine for quick demos. It becomes unreliable once you move to multi-step workflows and production traffic.</p><div style="background-color: #f8f9fa; padding: 24px; border-radius: 8px; border-left: 5px solid #e74c3c; margin-top: 16px;"><pre style="color: #f8f8f2; padding: 16px; border-radius: 5px; overflow-x: auto; font-size: 14px; margin: 0; border: 1px solid #555;"><code>prompt = "Extract invoice fields as JSON"response = llm.generate(prompt)# Works on day 1# Breaks later when format drifts or fields go missing</code></pre></div></div>

<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h2 style="font-size: 32px; color: #34495e; margin-bottom: 24px;">1.1: Common Failure Modes</h2><ul style="font-size: 17px; line-height: 2.0; list-style-position: inside; color: #2c3e50;"><li><strong>Output drift:</strong> response format changes across runs/models</li><li><strong>No strict contract:</strong> fields and data types are not guaranteed</li><li><strong>Prompt logic gets scattered:</strong> templates + helper code + parsers</li><li><strong>Debugging is slow:</strong> hard to trace where failures came from</li><li><strong>Scale-up cost rises:</strong> each new flow adds more glue code</li></ul><p style="font-size: 16px; color: #7f8c8d; line-height: 1.7; margin-top: 14px;">This is why teams outgrow raw prompt strings.</p></div>

---

<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 22px; color: #2c3e50;">1.2: Why PDL Came About</h2><div style="padding: 24px; background-color: #fadbd8; border-left: 5px solid #e74c3c; border-radius: 6px;"><p style="font-size: 17px; margin: 0 0 12px 0; line-height: 1.75; color: #2c3e50;">PDL was built to treat prompt workflows like real programs, not text snippets. The goal is clear logic, predictable execution, and easier maintenance.</p><p style="font-size: 16px; margin: 0; line-height: 1.75; color: #2c3e50;">In short: move from fragile string prompts to structured prompt programs.</p></div></div>

<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 22px; color: #2c3e50;">1.2: Design Targets for PDL</h2><div style="padding: 24px; background-color: #fef5e7; border-left: 5px solid #f39c12; border-radius: 6px;"><ul style="font-size: 16px; line-height: 1.95; margin: 0; padding-left: 22px; color: #2c3e50;"><li>Make prompt logic readable and reviewable</li><li>Add strict structured output checks (parser/spec)</li><li>Support loops, branching, variables, and reuse in one place</li><li>Cut down hidden behavior spread across app code</li></ul></div></div>

<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 22px; color: #2c3e50;">1.2: What Changes with PDL</h2><div style="padding: 24px; background-color: #d6eaf8; border-left: 5px solid #3498db; border-radius: 6px;"><ul style="font-size: 16px; line-height: 1.9; margin: 0; padding-left: 22px; color: #2c3e50;"><li>Before: prompt behavior hidden in strings and side scripts</li><li>After: prompt behavior visible as executable blocks</li><li>Before: parsing done manually after generation</li><li>After: parser/spec validates outputs up front</li></ul></div></div>

<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 18px; color: #2c3e50;">1.3: What PDL Is</h2><div style="padding: 24px; background-color: #eafaf1; border-radius: 8px; border-left: 5px solid #27ae60;"><p style="font-size: 16px; line-height: 1.8; margin: 0 0 12px 0; color: #2c3e50;">PDL is a declarative language for building LLM workflows. You define steps, logic, and validation as a single structured program.</p><p style="font-size: 16px; line-height: 1.8; margin: 0; color: #2c3e50;">It is built for teams that need stable behavior, not one-off prompt experiments.</p></div></div>

<div style="padding: 40px; max-width: 1040px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 18px; color: #2c3e50;">1.4: How PDL Fixes String Prompting Pain</h2><div style="padding: 24px; background-color: #f4ecf7; border-radius: 8px; border-left: 5px solid #8e44ad;"><ul style="font-size: 16px; line-height: 1.95; margin: 0; padding-left: 22px; color: #2c3e50;"><li>Prompt logic is explicit and reviewable</li><li>Validation is built in, not bolted on later</li><li>Multi-step workflows are easier to maintain</li><li>Reusability improves across projects and teams</li></ul></div></div>

<div style="padding: 40px; max-width: 1100px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.1: Three-Layer Architecture</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 30px;">This is the core concept of how PDL works.</p>

<div style="background-color: #f0f0f0; padding: 30px; border-radius: 8px; font-family: 'Courier New', monospace; font-size: 13px; line-height: 1.6; overflow-x: auto; margin-bottom: 35px; border: 1px solid #ddd;">
<pre style="margin: 0; text-align: center">┌─────────────────────────────────────────────────────────────────────┐
│ LAYER 3: LLM Call                                                   │
│ Uses accumulated context as prompt                                  │
└──────────────────────────────┬──────────────────────────────────────┘
                               ↑ Passes context
┌──────────────────────────────┴──────────────────────────────────────┐
│ LAYER 2: Background Context                                         │
│ Accumulates all previous outputs                                    │
│ [system, user, assistant, user, assistant...]                       │
└──────────────────────────────┬──────────────────────────────────────┘
                               ↑ Data flows up
┌──────────────────────────────┴──────────────────────────────────────┐
│ LAYER 1: Individual Blocks                                          │
│ text:, model:, if:, for:, defs: ...                                 │
│ Each block produces data or logic                                   │
└─────────────────────────────────────────────────────────────────────┘</pre>
</div>

<div style="display: grid; grid-template-columns: 2fr 1fr; gap: 30px;">
<div>
<h3 style="font-size: 22px; margin-bottom: 20px; color: #34495e;\">Execution Flow</h3>
<ol style="font-size: 18px; line-height: 2.2; margin: 0; padding-left: 25px;">
<li>Execute <code>text:</code> block → added to context</li>
<li>Execute <code>text:</code> block → added to context</li>
<li>Execute <code>model:</code> block → LLM sees <strong>everything above</strong></li>
<li>LLM response → added to context</li>
<li>Next block continues...</li>
</ol>
</div>
</div>
</div>

---

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.2: Understanding Context Accumulation</h2><div style="padding: 25px; background-color: #e8f4f8; border-radius: 8px; border-left: 5px solid #2980b9; margin-bottom: 25px;"><h3 style="font-size: 20px; margin-top: 0; margin-bottom: 12px; color: #165fa0;">What is Context Accumulation?</h3><p style="font-size: 16px; color: #1e5f9e; margin: 0; line-height: 1.7;"><strong>Context accumulation</strong> is the process where PDL automatically combines all previous text blocks and LLM responses into a single conversation history that gets sent to the next LLM call. Instead of you manually managing what to include in each prompt, PDL remembers everything and makes it available automatically.</p></div></div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.2: With and Without PDL</h2>
<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 25px; margin-bottom: 25px;">
<div style="padding: 20px; background-color: #fdeaea; border-radius: 8px; border-left: 4px solid #e74c3c;">
<h4 style="font-size: 16px; margin-top: 0; margin-bottom: 10px; color: #c0392b;">Without PDL (Manual)</h4>
<pre style="margin: 0; font-size: 12px; color: #2c3e50; background-color: #fff;">
<code>context = "You are helpful\n"
    response1 = llm(context)
    context += response1
    context += "Next question?"
    response2 = llm(context)  
    # Manual!
</code>
</pre>
</div>
<div style="padding: 20px; background-color: #eafaf1; border-radius: 8px; border-left: 4px solid #27ae60;">
<h4 style="font-size: 16px; margin-top: 0; margin-bottom: 10px; color: #196f3d;">With PDL (Automatic)</h4>
<pre style="margin: 0; font-size: 12px; color: #2c3e50; background-color: #fff;">
<code>text:- "You are helpful\n"- 
model: llm- "Next question?\n"- 
model: llm  
# Auto uses all!</code>
</pre>
</div>
</div>
<p style="font-size: 16px; color: #7f8c8d; margin-bottom: 0; line-height: 1.7;">
<strong>Use Case:</strong> 
This demonstrates how PDL automatically manages conversation state. You write blocks in order, and PDL ensures each LLM call sees everything that came before it. This eliminates a huge source of bugs and cognitive overhead.</p>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;"><h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.2: Code Demonstration - Context Accumulation</h2><p style="font-size: 17px; color: #7f8c8d; margin-bottom: 0; line-height: 1.6;">Run this PDL program to see context accumulation in practice across sequential text and model blocks.</p></div>

````python
%%pdl --reset-context
description: Context accumulation demo
text:
- "You are an expert Python programmer."
- "You help users write clean, efficient code."
- "User: Write a function to check if a number is prime. Please include comments.\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0.7
    max_tokens: 200
````

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.3: Demo - Conditional Logic</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Use Case:</strong> Control program flow based on conditions. Execute different PDL blocks depending on variable values, comparison results, or dynamic data. This allows your PDL programs to adapt behavior at runtime.</p>
</div>

````python
%%pdl --reset-context
description: Conditional logic with scoring

if: ${85 > 80}
then:
  text: "Excellent score! (85 > 80)"
else:
  text: "Needs improvement (85 <= 80)"
````

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 30px; color: #2c3e50;">3.3.5: Control Outputs with `defs` and `contribute`</h2>

<div style="padding: 25px; background-color: #fdeaea; border-radius: 8px; border-left: 5px solid #e74c3c; margin-bottom: 0;">
<p style="font-size: 17px; margin: 0; color: #7f8c8d; line-height: 1.7;"><strong>The Problem:</strong> Sometimes you need intermediate computations that should not clutter the final output or LLM context. You want hidden calculations, partial contributions, or temporary data.</p>
</div>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 30px; color: #2c3e50;">3.3.5: defs</h2>

<div style="padding: 20px; background-color: #f0f0f0; border-radius: 8px; border-left: 4px solid #3498db;">
<h4 style="font-size: 20px; margin-top: 0; margin-bottom: 12px; color: #2980b9;"><code>defs:</code></h4>
<p style="font-size: 16px; color: #7f8c8d; margin: 0; line-height: 1.7;">Define variables locally. They do not get added to context or the final result.</p>
</div>
</div>

<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.4: Control Flow - Loops</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Use Case:</strong> Repeat blocks of code multiple times. PDL supports both counted loops (repeat N times) and list iteration (for each item). Loops are useful for batch processing, generating multiple variations, or iterating over datasets.</p>
</div>

````python
%%pdl --reset-context
description: Counted repeat loop with join

text:
- "Process items:\n"
- repeat: "- Item ${i}\n"
  index: i
  maxIterations: 3
  join:
    with: ""
````

<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.4.5: For Loops - Iterating Over Lists</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Use Case:</strong> Known collection iteration. When you have a fixed list of items, use `for` loops to iterate over each element and execute the same block. This is more readable than `repeat` when working with explicit lists or datasets.</p>
</div>

````python
%%pdl --reset-context
description: For loop over a list

defs:
  languages:
    data: ["Python", "JavaScript", "Go"]

for:
  lang: ${languages}
repeat:
  text: "- ${lang}\n"
````

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.5: Template Variables - Dynamic Context Substitution</h2>

<div style="padding: 25px; background-color: #e8f4f8; border-radius: 8px; border-left: 5px solid #2980b9; margin-bottom: 25px;">
<h3 style="font-size: 20px; margin-top: 0; margin-bottom: 12px; color: #165fa0;">Why Do We Need This?</h3>
<p style="font-size: 16px; color: #1e5f9e; margin: 0; line-height: 1.7;">Many real-world workflows require multiple LLM steps where each step uses outputs from the previous step. Template variables let you extract structured data (like JSON fields), then use those values in follow-up prompts. This enables complex multi-step reasoning pipelines.</p>
</div>

<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Real-World Example:</strong> Generate code with metadata (language, purpose), then use that metadata to write documentation that matches the code style.</p>
</div>

````python
%%pdl --reset-context
description: Template variables with dynamic context substitution

text:
- "Generate a code snippet with metadata in JSON format.\n"
- "Return ONLY: {\"language\": \"...\", \"purpose\": \"...\"}\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0.3
    max_tokens: 30
  parser: json
  spec:
    language: string
    purpose: string
  def: meta
- "Now write example code:\n"
- "Language: ${meta.language}\n"
- "Purpose: ${meta.purpose}\n"
- "Include comments and best practices.\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0.7
    max_tokens: 800
````

---

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.6: Structured Output with Parser & Spec</h2>

<div style="padding: 25px; background-color: #e8f4f8; border-radius: 8px; border-left: 5px solid #2980b9; margin-bottom: 25px;">
<h3 style="font-size: 20px; margin-top: 0; margin-bottom: 12px; color: #165fa0;">Why Do We Need This?</h3>
<p style="font-size: 16px; color: #1e5f9e; margin: 0; line-height: 1.7;">Without structured output validation, LLM responses are unreliable. Models surrounded with extra text ("Here's the JSON..."), forget to include fields, or provide wrong data types (like returning a string when you need a number). Production systems need guarantees: "give me valid JSON with exactly these fields and these types, or fail early." PDL's parser: + spec: validates responses BEFORE your code uses them, preventing downstream parsing errors and type mismatches.</p>
</div>

<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Real-World Example:</strong> When generating structured data like user IDs (must be integers), email addresses (must match format), or product recommendations (must be from a fixed set), you can't afford parsing failures or type mismatches.</p>
</div>

````python
%%pdl --reset-context
description: Structured JSON output with validation
text:
- "Write a Python function to sort a list.\n"
- "Return ONLY valid JSON with: {\"code\": \"...\", \"explanation\": \"...\"}\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0.1
  parser: json
  spec:
    code: string
    explanation: string
  def: result
  contribute: []
- "Generated Code:\n```python\n${result.code}\n```\n"
````

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<div style="padding: 25px; background-color: #eafaf1; border-radius: 8px; border-left: 5px solid #27ae60;">
<h3 style="font-size: 20px; margin-top: 0; margin-bottom: 18px; color: #196f3d;">PDL Parsing & Validation Pipeline</h3>
<ol style="font-size: 16px; line-height: 2.2; margin: 0; padding-left: 25px; color: #2c3e50;">
<li><strong>Extract</strong> JSON from padded LLM output</li>
<li><strong>Validate</strong> all required fields are present</li>
<li><strong>Type-check</strong> each field (string, number, etc.)</li>
<li><strong>Verify</strong> enum constraints and patterns</li>
<li><strong>Return</strong> fully typed and validated object</li>
</ol>
</div>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">3.7: Functions - Define & Reuse</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Use Case:</strong> Encapsulate reusable LLM interaction patterns. Define a function once with parameters, then call it multiple times with different arguments. Each call gets current context and can use template variables. Perfect for code review functions, summarizers, or any repeating pattern.</p>
</div>

````python
%%pdl --reset-context
description: Reusable functions - define once, call multiple times

defs:
  review_code:
    function:
      code_snippet: string
    return:
      lastOf:
      - |
        You are a Python code reviewer. Analyze this code and output ONLY valid JSON with these exact fields:
        - "quality": integer from 1 to 10
        - "feedback": brief string explaining the quality rating
        
        Code to review:
        ${code_snippet}
        
        Output ONLY the JSON object, nothing else. Do not include markdown, code blocks, or any explanation.
        Example format: {"quality": 8, "feedback": "Clean and efficient"}
      - model: ollama/granite-code:3b
        parameters:
          temperature: 0.1
          max_tokens: 100
        parser: json
        spec:
          quality: number
          feedback: string

text:
- "=== CODE REVIEW #1: Inefficient Algorithm ===\n"
- call: ${ review_code }
  args:
    code_snippet: "def fib(n): return fib(n-1) + fib(n-2)"
  def: review_1
- "Quality Score: ${review_1.quality}/10\n"
- "Feedback: ${review_1.feedback}\n\n"
- "=== CODE REVIEW #2: Efficient Sort ===\n"
- call: ${ review_code }
  args:
    code_snippet: "def sort_list(arr): return sorted(arr)"
  def: review_2
- "Quality Score: ${review_2.quality}/10\n"
- "Feedback: ${review_2.feedback}\n"
````

<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">Demo: Variable Definition in Action</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>Use Case:</strong> Observe how `def:` and `defs:` work together. We define setup variables with `defs:`, then use `def:` to store LLM outputs for later reuse. This pattern lets you build multi-step workflows with clean variable scoping.</p>

````python
%%pdl --reset-context
description: Variable definition with def and defs blocks

defs:
  system_prompt: "You are a helpful Python expert."
  code_style: "clean and Pythonic"

text:
- "${system_prompt}\n"
- "Write a simple function with ${code_style} style.\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0.7
    max_tokens: 100
  def: generated_code
- "\n"
- "Generated code stored in variable:\n"
- "${generated_code}\n"
- "\n"
- "Now test it:\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0.3
    max_tokens: 100
  def: test_result
- "Test result:\n${test_result}\n"
````

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 30px; color: #2c3e50;">4.2: Output Control with <code>contribute:</code></h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 0; line-height: 1.6;"><strong>The Problem:</strong> When calling an LLM, do you want its output to: (1) clutter the final user-facing result? (2) be visible to subsequent LLM calls? (3) be hidden entirely? By default, everything gets added everywhere. The <code>contribute:</code> option gives you precise control.</p>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 24px; color: #2c3e50;">4.2: <code>contribute: []</code></h2>
<div style="padding: 22px; background-color: #fdeaea; border-radius: 8px; border-left: 4px solid #e74c3c;">
<p style="font-size: 16px; margin: 0; color: #7f8c8d; line-height: 1.7;"><strong>Hidden:</strong> Compute this block, but do not add output to context or final result. Perfect for internal calculations and temporary helper steps.</p>
</div>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 24px; color: #2c3e50;">4.2: <code>contribute: [context]</code></h2>
<div style="padding: 22px; background-color: #f4ecf7; border-radius: 8px; border-left: 4px solid #8e44ad;">
<p style="font-size: 16px; margin: 0; color: #7f8c8d; line-height: 1.7;"><strong>LLM Context Only:</strong> Keep this output available for later model calls, but hide it from the final user-facing result. Great for intermediate reasoning steps.</p>
</div>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 24px; color: #2c3e50;">4.2: <code>contribute: [result]</code></h2>
<div style="padding: 22px; background-color: #eafaf1; border-radius: 8px; border-left: 4px solid #27ae60;">
<p style="font-size: 16px; margin: 0; color: #7f8c8d; line-height: 1.7;"><strong>User Output Only:</strong> Show this content in the final result, but do not add it to model context for future steps.</p>
</div>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 15px; color: #2c3e50;">Demo: Output Control in Action</h2>
<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 0; line-height: 1.6;"><strong>Use Case:</strong> Run this standalone demo to see how each <code>contribute:</code> mode affects visibility in context and final output.</p>
</div>

````python
%%pdl --reset-context
description: Fast contribute demo for live presentation

defs:
  essay_text: "AI can boost writing speed, but human review is still required for quality and factual accuracy."

text:
- "Goal: show contribute output control with minimal latency.\n"
- "Source text: ${essay_text}\n"
- text: "INTERNAL_CHECK: short routing step"
  contribute: []
- model: ollama/granite-code:3b
  parameters:
    temperature: 0
    max_tokens: 100
  contribute: [context]
  def: recommendations
  input: "From the source text, produce exactly 2 short improvement bullets."
- model: ollama/granite-code:3b
  parameters:
    temperature: 0
    max_tokens: 100
  contribute: [result]
  input: "Using the hidden recommendations, output exactly one concise final sentence for end users. Do not output bullet points."
````

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 30px; color: #2c3e50;">4.3: Type Safety with <code>parser:</code> & <code>spec:</code></h2>

<p style="font-size: 17px; color: #7f8c8d; margin-bottom: 25px; line-height: 1.6;"><strong>The Challenge:</strong> LLMs often pad JSON output with extra text ("Sure! Here's the JSON:"), forget fields, or provide wrong types. When you need structured data like <code>{"user_id": 123, "email": "invalid"}</code>, you cannot rely on manual parsing. The <code>parser:</code> + <code>spec:</code> combination extracts and validates automatically.</p>

<h3 style="font-size: 20px; margin-bottom: 0; color: #34495e;">Before vs After: Without and With Validation</h3>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h4 style="font-size: 18px; margin-bottom: 12px; color: #7f8c8d; font-weight: 600;">WITHOUT Parser (Unreliable)</h4>

<pre style="color: #f8f8f2; padding: 20px; border-radius: 5px; font-size: 13px; overflow-x: auto; margin-bottom: 0; border: 1px solid #555;"><code>- model: ollama/granite-code:3b:latest
  text: "Extract user info from: John, john@example.com. Return JSON: {name, email}"
  
# Result might be:
# "Sure! Here's the info:\n{\n  \"name\": \"John\",\n  \"email\": \"john@example.com\"\n}\n\nHope that helps!"
# 
# Your code has to:
# 1. Find the JSON inside this messy output
# 2. Parse it manually  
# 3. Hope all expected fields are present
# 4. Handle type mismatches (e.g., email as number)</code></pre>
</div>

````python
%%pdl --reset-context
description: Parser and spec live demo

text:
- "Extract user info from: John, john@example.com. Return JSON only.\n"
- model: ollama/granite-code:3b
  parameters:
    temperature: 0
    max_tokens: 40
  parser: json
  spec:
    name: string
    email: string
  def: user_data
  contribute: []
- "Validated user object:\n${user_data}\n"
````

<p style="font-size: 13px; margin: 0; color: #7f8c8d;">Pull JSON from padded/messy output</p>

<div style="padding: 18px; background-color: #fef5e7; border-radius: 8px;">

<h2 style="font-size: 32px; margin-bottom: 30px; color: #2c3e50;">5.2: When to Use AutoPDL & Competitive Landscape</h2>

<h3 style="font-size: 22px; margin-bottom: 20px; color: #34495e;">How Does AutoPDL Compare to Other Approaches?</h3>

<div style="display: grid; grid-template-columns: 1fr 1fr 1fr; gap: 18px; margin-bottom: 0;">
<div style="padding: 18px; background-color: #ecf0f1; border-radius: 8px; border-left: 4px solid #7f8c8d;">
<p style="font-size: 13px; font-weight: bold; margin: 0 0 8px 0; color: #2c3e50;">Manual Tuning</p>
<p style="font-size: 13px; margin: 0; color: #7f8c8d;">Hand-adjust prompts, models, examples. Slow, subjective, hard to reproduce.</p>
</div>
<div style="padding: 18px; background-color: #ecf0f1; border-radius: 8px; border-left: 4px solid #7f8c8d;">
<p style="font-size: 13px; font-weight: bold; margin: 0 0 8px 0; color: #2c3e50;">DSPy Optimization</p>
<p style="font-size: 13px; margin: 0; color: #7f8c8d;">Framework for optimizing prompt components. Requires rewriting logic in DSPy.</p>
</div>
<div style="padding: 18px; background-color: #d5f4e6; border-radius: 8px; border-left: 4px solid #27ae60;">
<p style="font-size: 13px; font-weight: bold; margin: 0 0 8px 0; color: #196f3d;">AutoPDL (PDL)</p>
<p style="font-size: 13px; margin: 0; color: #27ae60;"><strong>Built-in optimization.</strong> No rewriting needed; your existing PDL just works. Declarative + automated.</p>
</div>
</div>
</div>

<div style="padding: 40px; max-width: 1000px; margin: 0 auto;">
<h2 style="font-size: 32px; margin-bottom: 30px; color: #2c3e50;">5.2: When to Use AutoPDL & Competitive Landscape</h2>

<h3 style="font-size: 22px; margin-bottom: 20px; color: #34495e;">When to Use AutoPDL</h3>

<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 25px;">
<div style="padding: 25px; background-color: #eafaf1; border-radius: 8px; border-left: 5px solid #27ae60;">
<h4 style="font-size: 18px; margin-top: 0; margin-bottom: 15px; color: #196f3d;">Use AutoPDL When:</h4>
<ul style="font-size: 16px; line-height: 2.1; margin: 0; padding-left: 25px; color: #2c3e50;">
<li>Working PDL program exists</li>
<li>Have labeled dataset available</li>
<li>Want to optimize performance</li>
<li>Can define a scoring metric</li>
</ul>
</div>

<div style="padding: 25px; background-color: #fdeaea; border-radius: 8px; border-left: 5px solid #e74c3c;">
<h4 style="font-size: 18px; margin-top: 0; margin-bottom: 15px; color: #c0392b;">Not Appropriate When:</h4>
<ul style="font-size: 16px; line-height: 2.1; margin: 0; padding-left: 25px; color: #2c3e50;">
<li>Still prototyping basics</li>
<li>No benchmark dataset</li>
<li>Exploring concepts</li>
<li>Constraints too loose</li>
</ul>
</div>
</div>
</div>
